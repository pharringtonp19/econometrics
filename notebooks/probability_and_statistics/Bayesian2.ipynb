```python
import jax 
import jax.numpy as jnp 
from oryx.core.ppl import random_variable
from tensorflow_probability.substrates import jax as tfp
tfd = tfp.distributions
import matplotlib.pyplot as plt
from functools import partial 
from oryx.core.ppl import log_prob
```

```python
def Sigma(key): 
  return random_variable(tfd.Exponential(rate=1.0))(key)

x = jnp.linspace(0., 5., 1000)
plt.plot(x, jnp.exp(jax.vmap(log_prob(Sigma))(x)))
plt.show()
```

```python
def Alpha_Bar(key): 
  return random_variable(tfd.Normal(loc=0., scale=1.5))(key)

x = jnp.linspace(-6., 6., 1000)
plt.plot(x, jnp.exp(jax.vmap(log_prob(Alpha_Bar))(x)))
plt.show()
```

```python
def Alpha_J(alpha_bar, sigma, key): 
  return random_variable(tfd.Normal(loc=alpha_bar, scale=sigma))(key)
```
